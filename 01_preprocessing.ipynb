# Combine Daily Temperature CSV Parts

This notebook joins the three raw daily temperature CSV parts from `data/raw/` into one pandas DataFrame, filters out non-depot cities, and writes the depot-only dataset to `data/processed/daily-temperature-for-depots.csv`.

```python
from pathlib import Path

import pandas as pd

PROJECT_DIR = Path.cwd()
RAW_DIR = PROJECT_DIR / "data" / "raw"
PROCESSED_DIR = PROJECT_DIR / "data" / "processed"
PROCESSED_DIR.mkdir(exist_ok=True)

pd.set_option("display.max_columns", 50)
```

```python
csv_files = [
    RAW_DIR / "daily-temperature-for-30-sites-to-2022-part1.csv",
    RAW_DIR / "daily-temperature-for-30-sites-to-2022-part2.csv",
    RAW_DIR / "daily-temperature-for-30-sites-to-2022-part3.csv",
]
```

```python
parts = [pd.read_csv(file) for file in csv_files]
df = pd.concat(parts, ignore_index=True)

print(f"Combined rows: {len(df):,}")
print(f"Combined columns: {df.shape[1]:,}")
df.head()
```

```python
df['site'].unique()
```

```python
sites_to_keep = ['Whangārei (Northland)','Auckland (Auckland)','Hamilton (Waikato)','Tauranga (Bay of Plenty)',"Napier (Hawke's Bay)",'New Plymouth (Taranaki)','Dannevirke (Manawatū-Whanganui)','Wellington (Wellington)','Blenheim (Marlborough)','Christchurch (Canterbury)','Queenstown (Otago)','Dunedin (Otago)']

filtered = df[df['site'].isin(sites_to_keep)]

filtered
```

```python
mapping = {'Whangārei (Northland)': 'Whangarei','Auckland (Auckland)': 'Auckland','Hamilton (Waikato)': 'Hamilton','Tauranga (Bay of Plenty)': 'Tauranga',"Napier (Hawke's Bay)": 'Napier','New Plymouth (Taranaki)': 'New Plymouth','Dannevirke (Manawatū-Whanganui)': 'Palmerston North','Wellington (Wellington)':'Wellington','Blenheim (Marlborough)':'Blenheim','Christchurch (Canterbury)':'Christchurch','Queenstown (Otago)':'Queenstown','Dunedin (Otago)':'Dunedin'}

filtered['site'] = filtered['site'].map(mapping)

filtered
```

```python
filtered.info()
filtered.tail()
```

```python
output_path = PROCESSED_DIR / "daily-temperature-for-depots.csv"
filtered.to_csv(output_path, index=False)

print(f"Saved combined CSV to: {output_path.resolve()}")
```
